```python
#funcao q recebe uma lista de numeros usando percentil

lista = [160, 168, 174, 180, 183, 188, 190, 198]

def percentil_ruim(lista, percentil):
    n = len(lista)
    posicao = (percentil / 100) * (n - 1)
    indice = round(posicao)
    return lista[indice]


print(percentil_ruim(lista, 50))


def percentil_alvo(lista, alvo):
    n = len(lista)
    contagem = 0
    for x in lista:
        if x <= alvo:
            contagem += 1

    return (contagem / n) * 100

print(f"O valor 174 está no percentil: {percentil_alvo(lista, 174)}%")




def percentil(lista, percentil):
    n = len(lista)

    posicao = (percentil / 100) * n

    if posicao % 1 == 0:
        indice = int(posicao) - 1
        return lista[indice]
    else:
        indice_baixo = int(posicao) - 1  # int(2.5) - 1 = 1
        indice_cima = int(posicao)      # int(2.5) = 2

        if indice_baixo < 0:
            indice_baixo = 0

        valor1 = lista[indice_baixo]
        valor2 = lista[indice_cima]
        return (valor1 + valor2) / 2

minha_lista = [160, 168, 174, 180, 183, 188, 190, 198]

# Teste 1: k = 25 (i = 0.25 * 8 = 2.0 -> Inteiro)
print(f"Percentil 25: {percentil(minha_lista, 25)}")

# Teste 2: k = 50 (i = 0.5 * 8 = 4.0 -> Inteiro)
print(f"Percentil 50: {percentil(minha_lista, 50)}")

# Teste 3: k = 30 (i = 0.3 * 8 = 2.4 -> Não inteiro)
print(f"Percentil 30: {percentil(minha_lista, 30)}")

```
