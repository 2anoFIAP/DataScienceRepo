```python
#funcao q recebe uma lista de numeros usando percentil


#TESTES
'''lista = [160, 168, 174, 180, 183, 188, 190, 198]

def percentil_ruim(lista, percentil):
    n = len(lista)
    posicao = (percentil / 100) * (n - 1)
    indice = round(posicao)
    return lista[indice]


print(percentil_ruim(lista, 50))


def percentil_alvo(lista, alvo):
    n = len(lista)
    contagem = 0
    for x in lista:
        if x <= alvo:
            contagem += 1

    return (contagem / n) * 100

print(f"O valor 174 está no percentil: {percentil_alvo(lista, 174)}%")
'''



def percentil(lista, percentil):
    n = len(lista) #percorre a lista e descobre a quantidade de elementos dentro

    posicao = (percentil / 100) * n # essa formula eu vi no google e apliquei aqui, basicamente ela pega o valor do percentil q a gente vai falar la no final (ex=50) ai divide por 100 e dps multiplica pelo numero de elementos q a gente calculou la na linhda de cima

    if posicao % 1 == 0: #aqui e aquele negocio q o danilo explicou de pegar o resto da divisao por zero, se for zero a gente cria a variavel indice e armazena o inteiro da posicao, se fosse 2 por exemplo o python entenderia o segundo numero pq comeca em zero, ent tu diminui 1 pra nao dar o resulktado errado
        indice = int(posicao) - 1
        return lista[indice]
    else:
        indice_baixo = int(posicao) - 1  # essas duas linhas fazem a os numero quebrados, se for diferente de zero naquele primeiro if, cai nesse else, o int posicao corta as casas decimais ent se for 2.5 vira 2, por isso a gente tira 1 e no outor nao
        indice_cima = int(posicao)    #pra pegar o valor maior e o menor pra fazer a media tendeu

        if indice_baixo < 0: # aqui e so uma verificacao pra nao dar erro, se o indici baixo for menor q zero daria erro , pq ficaria negativo, ent nao pode ser menor, ent se ele for, so iguala ele a zero
            indice_baixo = 0

        valor1 = lista[indice_baixo] # aqui nos atribui os valores e blz
        valor2 = lista[indice_cima]
        return (valor1 + valor2) / 2 # aqui vai ser o final das condicoes pq ele vai retornar a media

minha_lista = [160, 168, 174, 180, 183, 188, 190, 198] # a lista q o nemeczao passou

# Teste 1: k = 25 (i = 0.25 * 8 = 2.0 -> Inteiro)
print(f"Percentil 25: {percentil(minha_lista, 25)}") # testes inteiro (igual ta escrito em cima)

# Teste 2: k = 50 (i = 0.5 * 8 = 4.0 -> Inteiro)
print(f"Percentil 50: {percentil(minha_lista, 50)}") # testes inteiro (igual ta escrito em cima)

# Teste 3: k = 30 (i = 0.3 * 8 = 2.4 -> Não inteiro)
print(f"Percentil 30: {percentil(minha_lista, 30)}") # testes nao inteiro pra testar se ele faz a media

```
